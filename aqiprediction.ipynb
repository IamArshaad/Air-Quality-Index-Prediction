**Handling the missing valeus**



*   Convert -200 to NaN
*   Replace all NaN with mean of that specific column



```python
airquality_data = airquality_data.replace(to_replace=-200, value=np.NaN)
```

```python
#Check for missing values
airquality_data.isnull().sum()
```

```python

airquality_data.tail()
```

```python
airquality_data.select_dtypes(include='number').mean()
```

```python
#Replace the missing values with their mean valeus
airquality_data = airquality_data.fillna(airquality_data.select_dtypes(include='number').mean())
```

Two taska can be performed on this data set: one is forcasting and the other is regression.

**Forecasting with FB Prophet Algorithm**

FB Prophet documentation: https://facebook.github.io/prophet/docs/quick_start.html

```python

#Converting the date from DD/MM/YYYY to YYYY-MM-DD

date_info = pd.to_datetime(airquality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info=airquality_data['Time']
```

```python
#Change the time to proper format ('.' to ':')
time_info = time_info.apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#Combining 2 series into a pandas dataframe.
date_time = pd.concat([date_info,time_info],axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
#Combining date and time and creating a new column

date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

We have to convert ds from object data type to date time format.

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

Forecasting Relative humidity

```python
#Adding Temperature from the air quality data as athe y value
data['y'] = airquality_data['T']
```

```python
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet
```

```python
#Training the prophet model

model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
